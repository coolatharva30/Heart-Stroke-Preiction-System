```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('healthcare-dataset-stroke-data.csv')
df
```

```python
df.dropna(axis=1)
```

```python
df = df.drop_duplicates();
df
```

```python
#Encoding

ordinal_mapping = { 'Male': 1, 'Female': 0}
df['gender'] = df['gender'].map(ordinal_mapping)


ordinal_mapping = { 'formerly smoked': 0, 'smokes': 1,'never smoked':0}
df['smoking_status'] = df['smoking_status'].map(ordinal_mapping)

ordinal_mapping = { 'Rural': 0, 'Urban': 1}
df['Residence_type'] = df['Residence_type'].map(ordinal_mapping)

```

```python
df = df.dropna()
df
```

```python
x = df[['hypertension','heart_disease','Residence_type', 'age','avg_glucose_level','bmi','smoking_status']]
x
```

```python
upper_bound = q3 + (1.5 * iqr)
upper_bound
```
